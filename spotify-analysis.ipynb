```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_tracks = pd.read_csv("/Users/aarushi/Spotify Project/tracks.csv")
df_genre = pd.read_csv("/Users/aarushi/Spotify Project/SpotifyFeatures.csv")
```

```python
#checking for null values and the total number of null values

pd.isnull(df_tracks).sum()
```

```python
df_sorted = df_tracks.sort_values("popularity", ascending=True).head()
df_sorted
```

```python
df_tracks.describe().transpose()
```

```python
most_popular = df_tracks.query("popularity > 90", inplace=False).sort_values("popularity", ascending=False)
most_popular[0:10]
```

```python
df_tracks.set_index("release_date", inplace=True)
df_tracks.index = pd.to_datetime(df_tracks.index)
df_tracks.tail()
```

```python
df_tracks["duration"] = df_tracks["duration_ms"].apply(lambda x: round(x/1000))
df_tracks.drop("duration_ms", inplace=True, axis=1)
df_tracks.duration.head()
```

```python
corr_df = df_tracks.drop(["key", "mode", "explicit"], axis=1).corr(method="pearson")
plt.figure(figsize=(14,6))
heatmap = sns.heatmap(data=corr_df, annot=True, fmt=".1g", vmin=-1, vmax=1, center=0, cmap="inferno", linewidths=1, linecolor="Black")
heatmap.set_title("Correlations Heatmap between Variables")
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation = 90)

```

```python
#creating a sample of the original data to consider only 0.4% of it to make regression plots

sample_df = df_tracks.sample(int(0.004*len(df_tracks)))
print(len(sample_df))

```

```python
#plotting correlation between loudness and energy
#positive corr as per heatmap

plt.figure(figsize=(10,7))
sns.regplot(data=sample_df, x="energy", y="loudness", color="g").set(title="Loudness v/s Energy Correlation")
```

```python
#plotting correlation between popularity and acousticness
#negative corr as per heatmap

plt.figure(figsize=(10,7))
sns.regplot(data=sample_df, x="acousticness", y="popularity", color="b").set(title="Popularity v/s Acousticness Correlation")
```

```python
df_tracks["dates"] = df_tracks.index.get_level_values("release_date")
df_tracks.dates = pd.to_datetime(df_tracks.dates)
years = df_tracks.dates.dt.year
print(years)
years.min()
```

```python
#distribution plot
sns.displot(years, discrete=True, aspect=2, height=5, kind="hist").set(title="Number of songs per year")
```

```python
#change in duration of songs over the years
total_duration = df_tracks["duration"]
fig, ax = plt.subplots(figsize=(18,7))
fig = sns.barplot(x = years, y = total_duration, ax = ax, errwidth = False).set(title = "Year v/s Song Duration")
plt.xticks(rotation=90)
```

```python
#analyse which genre is the most popular

sns.color_palette("pastel", as_cmap=True)
sns.barplot(data=df_genre, y="genre", x="popularity")
plt.xlabel("Popularity")
plt.ylabel("Genres")
```

```python
# most popular artists
sns.set_style(style = "darkgrid")
plt.figure(figsize=(10,5))
popular = df_genre.sort_values("popularity", ascending=False).head(10)
sns.barplot(data=popular, y="artist_name", x="popularity").set(title="Top Artists by Popularity")
```
